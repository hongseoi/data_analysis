```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('C:/Users/HONG SEO I/Desktop/data_analysis/Web_crolling/chromedriver.exe')
```

```python
driver.implicitly_wait(3) #페이지 로드될 때까지 대기
driver.get("https://www.livesport.com/team/manchester-united/ppjDR086/") #원하는 페이지 불러오기
```

```python
#페이지 파싱
from bs4 import BeautifulSoup
```

```python
page = driver.page_source
```

```python
bs_obj = BeautifulSoup(page, "html.parser")
bs_obj
```

```python
win = bs_obj.find_all("span", {"class":"wld wld--w"})
win
```

```python
#정규시간 외 승리
tiewin = bs_obj.find_all("span",{"class":"wld wld --wo"})
tiewin
```

```python
#비긴경우
tie = bs_obj.find_all("span",{"class":"wld wld--d"})
tie
```

```python
#진경우
lose = bs_obj.find_all("span",{"class":"wld wld--l"})
lose
```

```python
#정규시간외패
tielose = bs_obj.find_all("span",{"class":"wld wld--lo"})
tielose
```

```python
n_win = len(win)+len(tiewin)
n_lose = len(lose)+len(tielose)
n_draw=len(tie)

print(n_win)
print(n_lose)
print(n_draw)
```

```python
# 승무패 중 가장 높은 수치를 기록하는 항목 찾기
report = {'win':n_win, 'draw':n_draw, 'lose':n_lose}
report
```

```python
max_n = max(report.values())
report.keys()
```

```python
for key in report:
    if report[key]==max_n:
        print(key)
```
