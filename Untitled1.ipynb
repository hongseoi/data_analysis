```python
import csv
import matplotlib.pyplot as plt

f = open('data\seoul.csv')
data = csv.reader(f)
next(data)
high = [] #최고 기온 저장할 리스트
low = [] #최저 기온 저장할 리스트

for row in data:
    if row[-1] != '' and row[-2] != '': #최고기온, 최저기온값이 존재한다면
        if 1983 <= int(row[0].split('-')[0]): #1983년 이후 데이터라면
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14': #2월 14일의 데이터라면
                high.append(float(row[-1]))
                low.append(float(row[-2]))
            

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False #마이너스 기호 깨짐 방지
plt.title("2월 14일의 기온 변화 그래프")

plt.plot(high, 'red', label = 'high')
plt.plot(low, 'blue', label = 'low')
plt.legend()  #범례 표시
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('data/seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    month = row[0].split('-')[1] #'1998-12-23' 형태의 row[0] 데이터
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

plt.hist(aug, bins=100, color = 'r', label = 'Aug')
plt.hist(jan, bins=100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('data/seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    month = row[0].split('-')[1] #'1998-12-23' 형태의 row[0] 데이터
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

plt.boxplot(aug)
plt.boxplot(jan)
plt.boxplpt([aug, jan]) #1월 8월의 최고기온 데이터를 원소로 하는 리스트를 boxplot 형태로 시각화
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('data/seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])  #day리스트 내 31개의 리스트 생성

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] =='08':
            
            #일별 최고 기온을 day리스트 내 리스트에 각각 저장
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot') #그래프 스타일 지정
plt.figure(figsize=(10,5), dpi=300) #그래프 크기 가로 10 세로 5로 수정
plt.boxplot(day, showfliers=False) #이상치 값 생략
plt.show()
```
