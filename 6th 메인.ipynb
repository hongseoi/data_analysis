```python
#numpy를 이용해 버블 차트 그리기
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,100,200)
y = np.random.randint(10,100,200)

size = np.random.rand(200)*100
                      
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
#ndarray: numpy의 특별한 데이터 타입의 n차원 배열 << 리스트 등 연속된 데이터
#리스트와의 차이점: 리스트내 쉽표구분 없음. 리스트처럼 인덱싱 가능

import numpy as np
a = np.array([1,2,3,4])
print(a)
print(a[1],a[-1])
print(a[1:])
```

```python
#numpy array는 한가지 타입의 데이터만 저장 가능. 
#3개의 정수와 1개의 문자가 저장된 리스트를 배열로 변환하면 모두 같은 문자 타입으로 치환됨

import numpy as np
a = np.array([1,2,'3',4])
print(a)
```

```python
#리스트 뿐만 아니라 zeros(), ones(), eye() 등 다양한 함수를 통해 배열 초기화 가능
import numpy as np
#0으로 이루어진 크기 10의 배열 생성
a = np.zeros(10)
print(a)
b = np.ones(10)
print(b)

#3행3열의 배열 생성
c = np.eye(3)
print(c)
```

```python
#연속된 숫자로 배열 생성
print(np.arange(3))
print(np.arange(3,7))
print(np.arange(3,7,2))

print(np.linspace(1,2,11))
```

```python
# numpy array의 활용
a = np.zeros(10) + 5
print(a)
```

```python
# 배열에 함수 적용 >> 배열의 모든 값이 계산됨
a = np.linspace(1,2,11)
print(np.sqrt(a))
```

```python
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))
plt.show()
```

```python
# mask: 조건에 부합하는 데이터만 선별적으로 저장
# 조건에 부합하면 True, 그렇지 않으면 False
a = np.arange(-5,5)
print(a)

#mask
print(a<0)

# 다시 배열에 적용하면 조건에 부합하는(True) 데이터만 남음
print(a[a<0])

# 마스크를 변수에 넣어서도 사용가능
# abs(a)>3 == 배열 a에 저장된 원소의 절대값이 보다 크다
mask = abs(a) > 3
print(a[mask])

##마스크를 연결해서도 사용 가능
mask1 = abs(a)>3
mask2 = abs(a)%2 == 0
print(a[mask1 + mask2]) # 두 조건 중 하나라도 참인경우
print(a[mask1 * mask2]) # 두 조건 모두 참일 경우
```

```python
# 버블차트에 마스크 적용하기
x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)

mask1 = abs(x)>50
mask2 = abs(y)>50
x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, s=100, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
