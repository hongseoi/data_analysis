```python
#2012~2022년의 소비자물가지수
import csv
import matplotlib.pyplot as plt

f= open('data/prc.csv')
data = csv.reader(f)

result = []
year = []

next(data)

for row in data:
    
    if row[2] != "":
        result.append(float(row[2]))
        
plt.title("2012~2022")
plt.plot(result, 'r')
plt.show()

    
```
