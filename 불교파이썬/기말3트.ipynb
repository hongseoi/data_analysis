```python
#1. 회귀분석
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import math
import seaborn as sns
import pandas as pd

df = pd.read_csv("src", encoding='cp949')

df.columns
df.head()

df['logx'] = df['x'].apply(math.log)

sns.distplot(df['x'])
sns.distplot(df['logx'])

y = df.logx
x = df.y

plt.scatter(y,x, alpha=0.5)

df_ols = smf.ols('y~logx', df).fit()
df_ols.summary()
```

```python
#5. bargraph
import turtle
t = turtle.Turtle()

def draw_bar(height):
    t.begin_fill()
    t.left(90)
    t.forward(height)
    t.write(str(height), font=("Times New Roman", 17, 'bold'))
    t.right(90)
    t.forward(40)
    t.right(90)
    t.forward(height)
    t.left(90)
    t.end_fill()
    
data = [1, 2, 3, 4, 5]
t.color("red")
t.fillcolor("blue")
t.pensize(3)

for d in data:
    draw_bar(d)
```

```python
#6. drawing taguk
import turtle
t = turtle.Turtle()
t.shape("turtle")

def draw_shape(radius,color1):
    t.left(270)
    t.width(3)
    t.color('black', color1)
    
    t.begin_fill()
    t.circle(radius/2.0, -180)
    t.circle(radius, 180)
    t.left(180)
    t.circle(-radius/2.0, -180)
    t.end_fill()


t.reset()
draw_shape(200, "red")
t.setheading(180)
draw_shape(200, "blue")
    
```

```python
#7. guess num
import random

tries = 0
guess = 0
num = random.randint(1,100)

while guess == num:
    guess = int(input("숫자 입력: "))
    tries += 1
    if guess > num:
        print("high")
    elif guess < num:
        print("low")
if guess == num:
    print("u match num in ", tries, " tries")
    
```

```python
#9. drawing with mouse
import turtle

t = turtle.Turtle()

def draw(x,y):
    t.goto(x,y)

t.pensize(10)
s = turtle.Screen()
s.onscreenclick(draw)
s.onkey(t.penup, "Up")
s.onkey(t.pendown, "Down")
s.listen()
```
