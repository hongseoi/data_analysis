```python
#2
import csv
f = open('subwayfee.csv')
data = csv.reader(f)

next(data)

pay = 0
unpay = 0
pay2 = 0
unpay2 = 0
total = []

for row in data:
    row[4:] = map(int, row[4:])
    pay += row[4]
    pay2 += row[6]
    unpay += row[5]
    unpay2 += row[7]
    
total.append(pay)
total.append(unpay)
total.append(pay2)
total.append(unpay2)
    
import matplotlib.pyplot as plt
plt.pie(total)
color = ['blue', 'orange','green','red']
plt.axis('equal')

plt.rc('font', family='Malgun Gothic')
plt.show()
```

```python
#3
import csv
f = open('subwayfee.csv')
data = csv.reader(f)

label = next(data)

pay = 0
unpay = 0
pay2 = 0
unpay2 = 0
total = []

for row in data:
    row[4:] = map(int, row[4:])
    pay += row[4]
    pay2 += row[6]
    unpay += row[5]
    unpay2 += row[7]
    
total.append(pay)
total.append(unpay)
total.append(pay2)
total.append(unpay2)
    
import matplotlib.pyplot as plt
plt.pie(total, autopct='%1.1f%%', labels=label[4:])
plt.title("서울역 1호선")
color = ['blue', 'orange','green','red']
plt.axis('equal')

plt.rc('font', family='Malgun Gothic')
plt.show()
```

```python
#4 호선을 입력받고 입력받은 호선의 유임승차 인원을 막대 그래프로 출력
import csv
f = open('subwayfee.csv', encoding='cp949')
data = csv.reader(f)

next(data)
subway = input("출력할 호선을 입력하세요")
people = []
x = []

for row in data:
    row[4:] = map(int, row[4:])
    if subway in row[1]:
        people.append(row[4])
        x.append(row[3])
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.bar(x,people, color='red')
plt.title("subway")
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()

```

```python
#5. 역의 이름과 호선을 입력받고, 입력받은 역의 유무임 승하차 인원을 막대 그래프로 출력
import csv
f = open('subwayfee.csv')
data = csv.reader(f)

label = next(data)
subway = input("출력할 호선과 역을 입력하세요:")
total = []
pay_in = 0
pay_out = 0
unpay_in = 0
unpay_out = 0

subway = subway.split()

for row in data:
    if subway[0] == row[1] and subway[1]==row[3]:
        pay_in += int(row[4])
        pay_out += int(row[5])
        unpay_in += int(row[6])
        unpay_out += int(row[7])  
        
total.append(pay_in)
total.append(pay_out)
total.append(unpay_in)
total.append(unpay_out)

print(total)

import matplotlib.pyplot as plt
plt.bar(label[4:], total, color='blue')
plt.rc('font', family='Malgun Gothic')
plt.title(subway)
plt.show()
```
