```python
#2
import csv
f = open('subwayfee.csv')
data = csv.reader(f)

next(data)

pay = 0
unpay = 0
pay2 = 0
unpay2 = 0
total = []

for row in data:
    row[4:] = map(int, row[4:])
    pay += row[4]
    pay2 += row[6]
    unpay += row[5]
    unpay2 += row[7]
    
total.append(pay)
total.append(unpay)
total.append(pay2)
total.append(unpay2)
    
import matplotlib.pyplot as plt
plt.pie(total)
color = ['blue', 'orange','green','red']
plt.axis('equal')

plt.rc('font', family='Malgun Gothic')
plt.show()
```

```python
#4 호선을 입력받고 입력받은 호선의 유임승차 인원을 막대 그래프로 출력
import csv
f = open('subwayfee.csv')
data = csv.reader(f)

next(data)
subway = input("출력할 호선을 입력하세요")
people = []
x = []

for row in data:
    row[4:] = map(int, row[4:])
    if subway in row[1]:
        people.append(row[4])
        x.append(row[3])
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.bar(x,people, color='red')
plt.title("subway")
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()

```

```python
#5. 역의 이름과 호선을 입력받고, 입력받은 역의 유무임 승하차 인원을 막대 그래프로 출력
import csv
f = open('subwayfee.csv')
data = csv.reader(f)

a = next(data)
subway = input("출력할 호선과 역을 입력하세요:")
lst = []

for row in data:
    if row[0] == subway:
        lst.append(int(row[4]))
        lst.append(int(row[5]))
        lst.append(int(row[6]))
        lst.append(int(row[7]))

        
import matplotlib.pyplot as plt
plt.bar(a[4:], lst, color='blue')
plt.title(subway)
plt.show()
```
