```python
#2
df = pd.DataFrame({
    'name':['A','B','C','D','E','F','G'],
    'horse power':[130,250,190,300,210,220,170],
    'weight':[1.9,2.6,2.2,2.9,2.4,2.3,2.2],
    'efficiency':[16.3,10.2,11.1,7.1,12.1,13.2,14.2]
})

df
```

```python
df = df.set_index('name')
df
```

```python
#2.2
df[df['horse power'] == df['horse power'].max()]
```

```python
#2.3
df['hp x mile'] = df['horse power'] * df['efficiency']
df
```

```python
df[df['hp x mile'] == df['hp x mile'].max()]
```

```python
#3
car = pd.DataFrame({
    'name':['A','B','C','D'],
    'horse power':[120,220,120,200],
    'weight':[1.9,2.1,1.5,2.9],
    'efficiency':[18.3,19.2,21.1,17.3]
})

car = car.set_index('name')
car
```

```python
#3.2
df_sum = pd.concat([df, car])
df_sum
```

```python
#3.3
df['com'] = 'P'
car['com'] = 'Q'

df_new = pd.concat([df,car])
df_new
```

```python
#3.4
df_new['hp x mile'] = df_new['horse power'] * df_new['efficiency']
df_new.groupby('com').mean()['hp x mile']
```
