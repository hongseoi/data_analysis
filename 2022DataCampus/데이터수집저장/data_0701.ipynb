```python
import numpy as np
import matplotlib.pyplot as plt

pure = np.linspace(1,10, 100) #1부터 10까지 100개의 데이터 생성
noise = np.random.normal(0,1,100) #평균이 0이고 표준편차가 1인 100개의 난수 생성

#넘파이 배열간 덧셈, 요소별 덧샘
signal = pure + noise

#그래프 그리기
plt.plot(signal)
plt.show()

```

```python
#MSE
import numpy as np

ypred = np.array([1,0,0,0,0])
y = np.array([0,1,0,0,0])

n=5
MSE = (1/n)*np.sum(np.square(ypred-y)) #square 제곱의미. 시그마는 np.sum
print(MSE)
```
