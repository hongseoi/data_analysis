```python
import csv

f=open('gender.csv')
data = csv.reader(f)

size = []
m = 0
f = 0
name = input("찾고싶은 지역명:")

for row in data:
    if name in row[0]:
        for i in range(101):
            m+=int(row[i+3])
            f+=int(row[i+106])

        break

size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')

plt.pie(size, labels=['남','여'], autopct='%.1f%%',colors=color, startangle=90)
plt.title(name+' 지역의 남녀 성별 비율')
plt.show()
```

```python
#아침 7시 승차 데이터 개수 및 인원수 출력하기

import csv
f = open('subwaytime.csv')
data = csv.reader(f)


next(data)
next(data)
result = []

for row in data:
    row[4:] = map(int, row[4:])
    result.append(sum(row[10:15:2]))

import matplotlib.pyplot as plt
result.sort()
plt.style.use('ggplot')
plt.bar(range(len(result)), height=result)
```

```python
#아침 7~9시 하차인원 최대역

import csv
f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
max_station = ''
max_people = 0

for row in data:
    row[4:] = map(int, row[4:])
    if sum(row[11:16:2]) > max_people:
        max_station = row[3] +'('+row[1]+')'
        max_people = sum(row[11:16:2])

print(max_station, max_people)
```

```python
#밤 11시에 사람들이 가장 많이 타는 역

import csv
f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
max_station = ''
max_people = 0

time = int(input("몇 시의 승차 인원이 가장 많은 역이 궁금하세요?: "))

for row in data:
    row[4:] = map(int, row[4:])
    a = row[4+(time-4)*2]
    if a>max_people:
        max_people = a
        max_station = row[3]+row[1]
print(max_station, max_people)
```
