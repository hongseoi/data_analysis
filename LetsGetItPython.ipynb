```python
import matplotlib.pyplot as plt
plt.plot([1,5,7,3,7])
plt.show()
```

```python
month = [3,4,5,6,7]
sales = [1,5,7,3,7]
plt.plot(month, sales)
plt.show()
```

```python
plt.rc('font', family="Malgun Gothic")
plt.title("월별 판매실적")
month = ["march","arpil","may","june",7]
sales = [1,5,7,3,7]
plt.plot(month, sales, color="y")
plt.show()
```

```python
#그래프 크기 바꾸기
plt.figure(dpi=150)
plt.rc('font', family='Malgun Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b')
plt.show()
```

```python
#범례 넣기
plt.rc('font', family='Malgun Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label='단위:개수')
plt.legend(loc='lower right')
plt.show()
```

```python
# 축 이름 넣기
plt.figure(dpi=100)
plt.rc('font', family='Malgun Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label="단위: 개수")
plt.legend(loc="lower left")
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```
