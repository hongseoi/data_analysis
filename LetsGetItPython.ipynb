```python
month = [3,4,5,6,7]
sales = [1,5,7,3,7]
plt.plot(month, sales)
plt.show()
```

```python
plt.rc('font', family="Malgun Gothic")
plt.title("월별 판매실적")
month = ["march","arpil","may","june",7]
sales = [1,5,7,3,7]
plt.plot(month, sales, color="y")
plt.show()
```

```python
#그래프 크기 바꾸기
plt.figure(dpi=150)
plt.rc('font', family='Malgun Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b')
plt.show()
```

```python
#범례 넣기
plt.rc('font', family='Malgun Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label='단위:개수')
plt.legend(loc='lower right')
plt.show()
```

```python
# 축 이름 넣기
plt.figure(dpi=100)
plt.rc('font', family='Malgun Gothic')
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label="단위: 개수")
plt.legend(loc="lower left")
plt.xlabel('월')
plt.ylabel('매출')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title("제목")  #그래프에 제목 넣기

plt.plot([10,20,30,40], color="blue", linestyle="--", label="asc")
plt.plot([40,30,20,10], "red", ls = ":", label="desc")
plt.legend() #그래프에 범례 넣기
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('data/seoul.csv')
data = csv.reader(f)
next(data)

max_temp = []

for row in data:
    if row[-1] != "":
        max_temp.append(float(row[-1]))

plt.plot(max_temp, 'r')
plt.show()
print(len(max_temp))
print(max_temp)
```

```python
#날짜 데이터 추출 by split()
data = '1990-02-13'
print(data.split('-'))
print(data.split('-')[0])
print(data.split('-')[1])
print(data.split('-')[2])
```

```python
# split 함수를 사용해서 8월의 최고기온 데이터 추출하고 그래프로 그리기

result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=='08':
            result.append(float(row[-1]))
plt.plot(result, 'hotpink')
plt.show()
            
```
