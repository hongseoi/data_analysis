```python
#drop
dfnd = pd.DataFrame(np.arange(16).reshape(4,4),
                    index=['ohio','colorado','utah','newyork'],
                    columns=['one','two','three','four'])

dfnd = dfnd.drop(['colorado','ohio'])
dfnd
```

```python
#axis 기본값은 0으로 기본적으로 행을 삭제. drop을 사용시 바로 적용되지는 않음
dfnd.drop('two',axis=1)
dfnd
```

```python
#axis 기본값은 0으로 기본적으로 행을 삭제. drop을 사용시 바로 적용되지는 않음
#inplace 사용하면 바로 적용됨
dfnd.drop('two',axis=1, inplace=True)
dfnd
```

#인덱싱과 슬라이싱
<기존 python ,numpy와의 차이점>
1. 인덱스 정보가 정수가 아니어도 됨
2. loc, iloc 이용 가능
3. 인덱스 정보(index=[])일경우 [3:5]는 3부터 5까지(5포함)


```python
#series에서의 인덱싱과 슬라이싱
ss = pd.Series(np.arange(4.), index=['a','b','c','d'])
print(ss['b'])
print(ss[1])
print(ss[:2])

print(ss[['b','a','d']])
print(ss[[1,3]])
print(ss[ss<2])
```

```python
print(ss['b':'c'])
```

```python
#슬라이싱 정보로 값 변경가능
ss['b':'c'] = 5
ss
```

```python
#데이터프레임에서의 indexing and slicing
df = pd.DataFrame(np.arange(16).reshape((4,4)),
                  index=['ohio','colorado','utah','newyork'],
                  columns=['one','two','three','four'])

df['two']
df[:2]
df[['three','one']]
```

```python
#행과 열로 지정도 가능 by 프로퍼티 loc[] 메소드 아님!
df.loc[['colorado','newyork'], 'one']
```

```python
df.loc[:'utah','two']
```

```python
#iloc 파이썬 문법인 0,1,2.... 사용하기 위한 프로퍼티. 기존 파이썬 문법대로 마지막 정보 미출력
ser = pd.Series(np.arange(3.))
print(ser.iloc[1])
```
